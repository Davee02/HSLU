```python
import numpy as np
import matplotlib.pyplot as plt
xfit = np.linspace(-1, 3.5)

for w1, w2, b in [(1.5,2,-3.5), (3,4,-7), (6,8,-14)]:
    plt.figure(figsize=(8,8))
    plt.margins(0,0)
    label = '{}x {} {}y {} {} = '.format(w1, '+' if w2 >= 0 else '-', abs(w2), '+' if b >= 0 else '-', abs(b))  
    # ... = -1
    plt.plot(xfit, (-w1 * xfit - b -1) / w2, label = label+'-1', linestyle='--', color='blue')
    # ... = 0
    plt.plot(xfit, (-w1 * xfit - b) / w2, label = label+'  0', color='red')
    # ... = 1
    plt.plot(xfit, (-w1 * xfit - b +1) / w2, label = label+' +1', linestyle='--', color='black')
    plt.axvline(x = 0, color='gray', linestyle='--')
    plt.axhline(y = 0, color='gray', linestyle='--')
    plt.xlim(-1, 3.5)
    plt.ylim(-1, 3.5)
    plt.legend()

plt.show()
```
