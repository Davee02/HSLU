```python
df = pd.read_excel("cars_for_gini.xlsx")
df.columns = ["name", "horsepower", "seats", "color", "sold"]
df.head(15)
```

```python
def calculate_gini_impurity(data, feature_col, target_col):
    feature_values = data[feature_col].unique()
    target_values = data[target_col].unique()

    gini_impurity = 0
    for feature_value in feature_values:
        entries_matching_feature_value = data[data[feature_col] == feature_value]
        distribution = len(entries_matching_feature_value) / len(data)
        gini = 1
        for target_value in target_values:
            entries_matching_feature_value_and_target_value = entries_matching_feature_value.loc[(data[target_col] == target_value)]
            p = len(entries_matching_feature_value_and_target_value) / len(entries_matching_feature_value)
            gini -= p ** 2
        gini_impurity += distribution * gini

    return gini_impurity
```

```python
calculate_gini_impurity(df, "color", "sold")
```

```python
df["seats_splitted"] = pd.cut(df["seats"], bins=[0, 3, 4.5], labels=["0-3", "3-4.5"])
df["seats_splitted"].cat.add_categories("4.5+", inplace=True)
df["seats_splitted"].fillna("4.5+", inplace=True)
df.head(15)
```

```python
calculate_gini_impurity(df, "seats_splitted", "sold")
```

```python
df = df[df["seats"] > 4.5]
calculate_gini_impurity(df, "color", "sold")
```
