```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./API_SP.DYN.LE00.IN_DS2_en_csv_v2_5994853.csv')
data.head()
```

```python
years = data.columns[6:-2]
life_expectancy_data = data[years]

plot_data = life_expectancy_data.T
plot_data.head()
```

```python
average_life_expectancy = plot_data.mean(axis=1)
average_life_expectancy.head()
```

```python
country_names = data['Country Name']
country_names.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

countries_to_highlight = ["Switzerland", "Mali", "Iceland", "Chad", "China"]

for country in range(plot_data.shape[1]):
    if country_names[country] in countries_to_highlight:
        ax.plot(years, plot_data.iloc[:, country], label=country_names[country], color="blue", linewidth=1, zorder=99)
    else:
        ax.plot(years, plot_data.iloc[:, country], label=country_names[country], color="lightgrey", alpha=0.5)

ax.plot(years, average_life_expectancy, color='red', linewidth=3, label='Average Life Expectancy')

ax.set_title('Life Expectancy Over Time (1962-2021) for All Countries')
ax.set_xlabel('Year')
ax.set_ylim(ymin=0)
ax.set_ylabel('Life Expectancy (Years)')
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

```python
# Calculating the absolute and relative increases in life expectancy for each country
life_expectancy_change = plot_data.iloc[-1] - plot_data.iloc[0]  # Absolute change (Life expectancy in 2021 - 1990)
relative_life_expectancy_change = (life_expectancy_change / plot_data.iloc[0]) * 100  # Relative change in percentage
```

```python
# Identifying the countries with the largest absolute and relative increases
country_with_largest_absolute_increase = country_names[life_expectancy_change.idxmax()]
largest_absolute_increase_value = life_expectancy_change.max()

country_with_largest_relative_increase = country_names[relative_life_expectancy_change.idxmax()]
largest_relative_increase_value = relative_life_expectancy_change.max()

country_with_largest_absolute_increase, largest_absolute_increase_value, country_with_largest_relative_increase, largest_relative_increase_value
```

```python
average_absolute_increase = life_expectancy_change.mean()
average_relative_increase = relative_life_expectancy_change.mean()
```

```python
relative_increase_df = pd.DataFrame({
    'Country': country_names,
    'Relative Increase (%)': relative_life_expectancy_change
})
sorted_relative_increase_df = relative_increase_df.sort_values(by='Relative Increase (%)', ascending=False).head(30)

fig, ax = plt.subplots(figsize=(30, 5))
ax.bar(sorted_relative_increase_df['Country'], sorted_relative_increase_df['Relative Increase (%)'], color='skyblue')
ax.bar(["Average"], [average_relative_increase], color='red')

ax.set_title('Top 30 Countries by Relative Increase in Life Expectancy (1962-2021)')
ax.set_xlabel('Country')
ax.set_ylabel('Relative Increase in Life Expectancy (%)')
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.grid(axis='y')
```

```python
absolute_increase_df = pd.DataFrame({
    'Country': country_names,
    'Absolute Increase (Years)': life_expectancy_change
})
sorted_absolute_increase_df = absolute_increase_df.sort_values(by='Absolute Increase (Years)', ascending=False).head(30)

fig, ax = plt.subplots(figsize=(30, 5))
ax.bar(sorted_absolute_increase_df['Country'], sorted_absolute_increase_df['Absolute Increase (Years)'], color='skyblue')
ax.bar(["Average"], [average_absolute_increase], color='red')

ax.set_title('Top 30 Countries by Absolute Increase in Life Expectancy (1962-2021)')
ax.set_xlabel('Country')
ax.set_ylabel('Absolute Increase in Life Expectancy (Years)')
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.grid(axis='y')
```
