```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./API_SP.DYN.IMRT.IN_DS2_en_csv_v2_5996676.csv')
data.head()
```

```python
years = data.columns[4:-2]
child_mortality_data = data[years]

plot_data = child_mortality_data.T
plot_data.head()
```

```python
average_child_mortality = plot_data.mean(axis=1)
average_child_mortality.head()
```

```python
country_names = data['Country Name']
country_names.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

countries_to_highlight = ["Switzerland", "Mali", "Iceland", "Chad", "China"]

for country in range(plot_data.shape[1]):
    if country_names[country] in countries_to_highlight:
        ax.plot(years, plot_data.iloc[:, country], label=country_names[country], color="blue", linewidth=1, zorder=99)
    else:
        ax.plot(years, plot_data.iloc[:, country], label=country_names[country], color="lightgrey", alpha=0.5)

ax.plot(years, average_child_mortality, color='red', linewidth=3, label='Average Child Mortality')

fig.suptitle('Child mortality rate Over Time (1990-2021) for All Countries')
ax.set_xlabel('Year')
ax.set_ylim(ymin=0)
ax.set_ylabel('Child Mortality (per 1000 live births)')
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```
