```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./renewable-share-energy.csv')
data.head()
```

```python
world_data = data.query("Entity == 'World'").query("Year >= 1990")
world_data.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(world_data["Year"], world_data['Renewables (% equivalent primary energy)'], label="Renewables (% equivalent primary energy)", marker="o")

ax.set_xlabel('Year')
ax.set_ylim(ymin=0)
ax.legend()
ax.grid()
ax.set_yticks(ax.get_yticks())
ax.set_yticklabels(['{:.1f}%'.format(x) for x in ax.get_yticks()])
```
