### Norms and binned colormaps

This was a continuous cmap. Let's get a binned one: 

````python
my_cmap_binned = plt.cm.get_cmap('coolwarm', 6)
````

````python
my_cmap_binned
````

````python
f, ax = plt.subplots()
p = ax.imshow(data_2d, cmap=my_cmap_binned)
f.colorbar(mappable=p)
````

Matplotlib has aggregated our data into bins. 

However: Our data goes -1..1, but the cmap only from 0..1, so there must be additional normalization there 

Without such a normalization, we would have had the following: 

````python
for x in [data_2d.min(), 0, data_2d.max()]:
    print("x: {:.4f}, cmap(x): {}".format(x,my_cmap_binned(x)))
````

We can access the normalization directly in imshow using `vmax` and `vmin`:

````python
f, ax = plt.subplots()
p = ax.imshow(data_2d, cmap=my_cmap_binned, vmax=0.5, vmin=-0.75)
f.colorbar(mappable=p)
````

Or we can use a separate norm: 

````python
norm = mpl.colors.Normalize(vmin=-1, vmax=1)
````

Let's check how it works: 

````python
for x in [-2, -1, 0, 1, 2]:
    print("x: {:.1f}, norm(x): {}".format(x,norm(x)))
````

We couple together the norm and cmap by calling `my_cmap(my_norm(x))`:

````python
for x in [data_2d.min(), 0, data_2d.max()]:
    print("x: {:.1f}, cmap(x): {}, cmap(norm(x)): {}".format(x,my_cmap_binned(x), my_cmap_binned(norm(x))))
````

So now it works as it should. 

This is a linear norm and by default it is happening in the background. If we want a non-linear transformation (e.g. log, sqrt, x^y), we have 2 choices: 
- manipulate the data values by hand and use a linear norm 
- leave data as it is and use a non-linear norm 

Matplotlib has a selection of in-built norms: https://matplotlib.org/stable/api/colors_api.html#color-norms 

More on defining norms: https://matplotlib.org/stable/tutorials/colors/colormapnorms.html

### Custom colormap with custom boundary norms where the colors should change

Let's customize it further. Assume that we want custom data ranges for the binned colors, so that the middle black stripe is only between -0.15 and 0.15:

````python
from matplotlib.colors import ListedColormap, BoundaryNorm

custom_cmap2 = ListedColormap(name="another_cmap", colors=['green', 'black', 'purple'])
norm2 = BoundaryNorm([-1, -0.15, 0.15, 1], custom_cmap2.N) 

custom_cmap2
````

Let's check how it works: 

````python
# BoundaryNorm maps values to INTEGERS instead of floats 0..1

for x in [-10, -2.5, -1.5, -1, -0.2, 0, 0.75, 0.9, 1.2, 2, 10]:
    print("x: {:.1f}, norm2(x): {}".format(x,norm2(x)))
````

So the mapping we get is:   
```
-inf to -1 => -1  "under"
-1 to -0.15 => 0  
-0.15 to 0.15 => 1  
0.15 to 1 => 2  
1 to +inf => 3  "over"
```

````python
# without the norm we get an equidistant color mapping that has boundaries at -0.33 and +0.33 

f, ax = plt.subplots()
p = ax.imshow(data_2d, cmap=custom_cmap2, )
f.colorbar(mappable=p, spacing='proportional') 
````

````python
# with the norm: 

f, ax = plt.subplots()
p = ax.imshow(data_2d, cmap=custom_cmap2, norm=norm2)
f.colorbar(mappable=p, spacing='proportional') # the `spacing` argument corrects the colorbar tick positions 
````

### Merge 2 colormaps: 

````python
cmap1 = LinearSegmentedColormap.from_list(name="cmap1", colors=["green", "black", "purple"])
cmap2 = LinearSegmentedColormap.from_list(name="cmap2", colors=["white", "yellow", "orange"])
````

````python
cmap1
````

````python
cmap2
````

Option 1: merge the two cmaps back-to-back, without blending: We need to generate such a long color list at the input that mpl will not be able to blend them.

````python
colors_cmap1 = cmap1(np.linspace(0, 1, 85)) # 85 = 1/3 * 256
colors_cmap2 = cmap2(np.linspace(0, 1, 171)) # 171 = 2/3 * 256
all_colors = np.vstack((colors_cmap1, colors_cmap2))

cmap_merged = LinearSegmentedColormap.from_list('merged',all_colors)

cmap_merged
````

Option 2: merge the colormaps with blending between them - generate only few colors and let mpl blend them again

````python
colors_cmap1 = cmap1(np.linspace(0, 1, 3))
colors_cmap2 = cmap2(np.linspace(0, 1, 6))
all_colors = np.vstack((colors_cmap1, colors_cmap2))

cmap_merged = LinearSegmentedColormap.from_list('merged',all_colors)

cmap_merged
````

Watch out: the 2 options are not 1:1 identical when it comes to positions of "pure" colors. Compare where the pure violet is in these 2 scenarios!

# Bonus: How to use a normalization with a cutoff in a seaborn plot 

Why it's nice to have a cutoff -> outstanding max values, we are only interested in variability in a particular range etc. 

````python
penguins.head()
````

Let's use a `relplot` to display the species data side-by-side:

````python
sns.relplot(x='bill_length_mm', col="species", y="bill_depth_mm", data=penguins, 
            hue='body_mass_g')
````

Let's use a LogNorm to make the points differ in color more: 

````python
from matplotlib.colors import LogNorm

sns.relplot(x='bill_length_mm', col="species", y="bill_depth_mm", data=penguins, 
            hue='body_mass_g', 
            hue_norm=LogNorm(vmin=penguins['body_mass_g'].min(), vmax=penguins['body_mass_g'].max()), 
            legend=True)
````

Why lognorm makes sense? inspect the distribution of the variable:

````python
sns.histplot(data=penguins, x='body_mass_g')
````

Compare the transformed distribution: 

````python
sns.histplot(LogNorm(vmin=penguins['body_mass_g'].min(), vmax=penguins['body_mass_g'].max())(penguins['body_mass_g']))
````

Finaly, apply the normalization with a cutoff:

````python
hue_cutoff = 4500
norm_hue = mpl.colors.Normalize(vmin=penguins['body_mass_g'].min(), vmax=hue_cutoff)

sns.relplot(x='bill_length_mm', col="species", y="bill_depth_mm", data=penguins, 
            hue='body_mass_g', hue_norm=norm_hue, 
            legend=True)
````

Unfortunately the hue normalization is not implemented in all of the charts... So sometimes you will need to go low-level again!
