# Vektor plotten

```python
def Vec (x, y):
    return np.array ([[x], [y]])

def PlotVec(p1, p2, color = "blue"):
    d = p2 - p1
    pl.arrow(p1[0, 0], p1[1, 0], d[0, 0], p2[1, 0] - p1[1, 0], head_width = 1, color = color, length_includes_head = True)

a = Vec (-7, -3)
b = Vec (13, -5)

pl.plot (x, y, 'o', color = 'red')
pl.axis ('equal')
PlotVec(a, b, "black")
pl.grid ()
```

# Vektoren addieren

```python
o = Vec (0, 0)
b = Vec(5, 0)
c = Vec(0, 10)

pl.axis('equal')
PlotVec(o, b)
PlotVec(o, c)
PlotVec(o, b + c, "red")

pl.grid()
```

# Vektor skalar multiplizieren

```python
a = Vec(0, 0)
b = Vec(5, 3)

pl.axis('equal')
PlotVec(a, b)
PlotVec(a, b * 2, "red")

pl.grid()
```

# Lineare Gleichungssysteme lsen

In der Form:

| 2a - 1b = 7  | </br>
| 2a + 2b = 1 |

Resultat:
a = 2.5, b = -2

```python
A = np.array ([
    [2, -1],
    [2, 2]
])

b = np.array ([
    [7],
    [1]
])

np.linalg.solve (A, b)

```

# Schnittwinkel zweier Vektoren

```python
u = np.array([4, 2, -1])  # Richtungsvektor erste Gerade
v = np.array([5, -2, 3])  # Richtungsvektor zweite Gerade
c = np.dot(u, v)/np.linalg.norm(u)/np.linalg.norm(v)  # -> cosine of the angle
angle = np.arccos(np.clip(c, -1, 1))  # if you really want the angle
np.rad2deg(angle)
```

# LU Zerlegung

```python
A = np.array ([
    [0, 1, 1],
    [1, 2, 1],
    [2, 7, 9]])

b = np.array ([
    [0],
    [0],
    [4]])
```

```python
# 1. und 3. Zeile tauschen
P = np.array ([
    [0, 0, 1],
    [0, 1, 0],
    [1, 0, 0]])

A = P @ A
v = P @ b

print(A)
print(b)
```

```python
E21 = np.array ([
    [1, 0, 0],
    [-0.5, 1, 0],
    [0, 0, 1]])

E21 @ A
```

```python
E32 = np.array ([
    [1, 0, 0],
    [0, 1, 0],
    [0, 2/3, 1]])

np.round(E32 @ E21 @ A, 5)
```

```python
R = E32 @ E21 @ A
L = np.linalg.inv(E21) @ np.linalg.inv(E32)

L
```

```python
# Kontrolle, ergibt wieder die Anfangsmatrix
L @ R
```

# Gauss-Jordan Verfahren

```python
M = np.array([
    [6, -1, -1, 4],
    [1, 1, 10, -6],
    [2, -1, 1, -2]
]).astype(float)
```

```python
M[1] = M[1] - M[0] / 6
M
```

```python
M[2] = M[2] - M[0] / 3
M
```

```python
M[2] = M[2] - M[2, 1] / M[1, 1] * M[1]
np.round(M, 5)
```

```python
M[1] = M[1] - M[1, 2] / M[2, 2] * M[2]
np.round(M, 5)
```

```python
M[0] = M[0] - M[0, 2] / M[2, 2] * M[2]
np.round(M, 5)
```

```python
M[0] = M[0] - M[0, 1] / M[1, 1] * M[1]
np.round(M, 5)
```

```python
# Einheitsmatrix erstellen
M[0] = M[0] / M[0, 0]
M[1] = M[1] / M[1, 1]
M[2] = M[2] / M[2, 2]

np.round(M, 5)
```
